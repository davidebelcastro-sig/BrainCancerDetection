# Get the mask for the image

```python
def fun(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.GaussianBlur(img, (3, 3), 0)
    img[img > 50] = 255
    img[img <= 50] = 0
    contours, hierarchy = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    copia = img.copy()
    ls_contorni = []
    for cnt in contours:
        cnt = cv2.convexHull(cnt)
        if cv2.contourArea(cnt) > 60:
            ls_contorni.append(cnt)
    
    ls_contorni = np.concatenate(ls_contorni)
    #convex
    ls_contorni = cv2.convexHull(ls_contorni)
    cv2.drawContours(copia, [ls_contorni], -1, 127, -1)
    copia[copia != 127] = 0
    copia[copia == 127] = 255
    #find contours
    contours, hierarchy = cv2.findContours(copia, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    mask = np.zeros(img.shape, np.uint8)
    cv2.drawContours(mask, contours, -1, 255, 1)
    return mask
```

# 3D view of the tumor

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
lista_pixel = []
#trova il contorno del rettangolo
dist = 1
start = 100
for k in range(1,97):
        img = fun('/Users/lucian/Documents/GitHub/BrainCancerDetection/notebooks/data/images/'+str(k)+'.png')
        for i in range(len(img)):
            for j in range(len(img[i])):
                if img[i][j] == 255:
                    lista_pixel.append([i,j,start])
        start -= dist
for el in lista_pixel:
        ax.scatter(el[0],el[1],el[2],c='gray')
        

    
ax.set_xlabel('X Label')
#scrivo su assey
ax.set_ylabel('Y Label')
#scrivo su assez
ax.set_zlabel('Z Label')
#setto lunghezza asse z

plt.show()
```
